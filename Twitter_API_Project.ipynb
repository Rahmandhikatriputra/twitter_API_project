```python
# Import important libraries
import snscrape.modules.twitter as sntwitter
import pandas as pd
import psycopg2
import datetime
from datetime import datetime as dt
```

#  Extract

```python
# Extract tweets using snscrape module
tweets = []

# Perform a for loop to extract the data
for tweet in list(sntwitter.TwitterSearchScraper('(from:ptjasamarga) until:2022-09-07 since:2022-09-04').get_items()):
    if len(tweets) == 500000:
        break
        
# We will only extract the tweet date, username account, the content and the hashtag on each tweet
    tweets.append([tweet.date, tweet.user.username, tweet.content, tweet.hashtags])
    
# Store the data in a dataframe format
tweets_df = pd.DataFrame(tweets, columns=['Datetime', 'Username', 'Tweet', 'Hashtags'])

df = tweets_df.copy()

df.head()
```

# Transform

```python
# Drop all rows that contain null values
df = df.dropna()

# Unzip all the list data in the Hashtags column
df["Hashtag"] = df["Hashtags"].apply(lambda x : ",".join(x))

# Create a new column contains actual time when the tweet has been uploaded
# Add the time data in Datetime column by 419 minutes
df["Actual Datetime Data"] = df["Datetime"].apply(lambda x : x + datetime.timedelta(minutes = 419))

# Structure the dataset
df = df[["Actual Datetime Data", "Username", "Tweet", "Hashtag"]].reset_index()

df.head()
```

# Load

```python
# Connect to the database using psycopg2
conn = psycopg2.connect(database="twitter_project",
                        user='postgres', password='Dika100196', 
                        host='localhost', port='5432')

# Create a cursor object
cursor = conn.cursor()

# First create a table inside the database

create_db = """
            CREATE TABLE tweets (
            ID int primary key, 
            Datetime timestamp,
            Username varchar (20),
            Tweet text,
            Hashtag varchar (30)
            );
            """

cursor.execute(create_db)

# Then insert all data into the database
insert_data = "INSERT INTO tweets VALUES (%s, %s, %s, %s, %s)"

for value in df.values:
    x = list(value)
    cursor.execute(insert_data, (x[0], x[1], x[2], x[3], x[4]))

conn.commit()
    
cursor.close()
conn.close()
```
